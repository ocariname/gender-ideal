```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
recruit = pd.read_csv('C:/Users/Admin/Desktop/gender-ideal-main/data_original/24_Recruitment_compensation_anon.csv')
pd.options.display.max_columns = None

# Original file had 13 columns, which excluded the other 11 companies. I used '24_Recruitment...'
# instead of 'Recruitment.csv', since the former CSV file includes additional data for Company 1N.
```

```python
# Reverse indices so that Company 1A is listed first.
recruit.index = recruit.index.values[::-1]
recruit = recruit[::-1]

# Insert 'Organization' at the beginning of the dataframe.
org_values = recruit.pop('Organization')
recruit.insert(0, 'Organization', org_values)
recruit
```

```python
# Remove text-based answer columns
# Q12, QD (?), 22 removed
recruit = recruit.drop(columns=["Recruitment, Promotion & Pipeline:  Tell us more about how your organization has tracked performance review results.",
                                "Give us feedback - how challenging is it to provide internal promotion details for your workforce to the degree asked above?",
                                "Recruitment, Promotion & Pipeline:  If you responded yes or in development, tell us more about the  program to retain and promote gender diverse entry and mid-level workers through training, development and other professional development opportunities.",
                                "Recruitment, Promotion & Pipeline: Before we move to the next section, is there anything else you want to tell us about how your organization tracks, reports on or is working to improve its recruitment, promotion and pipeline development?"], axis=1)
```

Q1: If "No" or "In progress" selected, following question will have no response.

```python
recruit.columns.values[1] = "Q1"
recruit.columns.values[2] = "Q2"

def determine_Q2(row):
    for row_index in range(recruit.shape[0]):
        q1 = recruit.iloc[row_index, 1]
        if q1 == "No" or q1 == "In progress":
            recruit.iloc[row_index, 2] = "N/A - No review conducted yet"

# axis = 0 refers to apply acting on rows
recruit.apply(determine_Q2, axis=0)    
```

Q3: If "No process has been designed yet", following question Q4 will have no response.

```python
recruit.columns.values[3] = "Q3"

def determine_Q3(row):
    for row_index in range(recruit.shape[0]):
        q1 = recruit.iloc[row_index, 3]
        if q1 == "No process has been designed yet":
            recruit.iloc[row_index, 4] = "N/A"
            recruit.iloc[row_index, 5] = "N/A"
            recruit.iloc[row_index, 6] = "N/A"

# axis = 0 refers to apply acting on rows
recruit.apply(determine_Q3, axis=0)    
```

```python
# Renaming columns based on Question # followed by _A/B/C/etc if a multiple selection
# Question numbering is from the survey at: https://gender-ideal.org/the-assessment
# Q4 is a choose all that apply question so each answer results in its own column
recruit.rename(columns={'Statement about gender equity goals in recruitment on all job postings':'Q4_A',
    'Process in-place (with designated and trained teammates or usage of technology) to review and revise job descriptions prior to posting to ensure gender-biased language is eliminated':'Q4_B',
    'Move toward skills-based competencies and growth mindsets and away from focus on certain schools, prior experiences':'Q4_C',
    'Requirement to post jobs on a diverse group of job boards, recruiting firms, etc to ensure opportunities are accessible to a diverse candidate pool':'Q4_D',
    'Blind resume review and blind auditions (as applicable)':'Q4_E',
    'Training for HR and all hiring managers on how to identify and reduce gender equity biases throughout the recruitment process':'Q4_F',
    'Oversight and support by trained HR or other designated-internal teams to ensure unbiased recruitment process':'Q4_G',
    'Oversight by a third-party entity to identify and reduce bias from recruitment process':'Q4_H',
                   'Other':'Q4_I'
                  },inplace=True)
```

Q5: If "No" or "In progress" selected, following question Q6 will have no response.

```python
recruit.columns.values[13] = "Q5"
recruit.columns.values[14] = "Q6"


def determine_Q5(row):
    for row_index in range(recruit.shape[0]):
        q5 = recruit.iloc[row_index, 13]
        if q5 == "No" or q5 == "In progress":
            recruit.iloc[row_index, 14] = "N/A - No targets set"
        # if no response, code the following question (Q6) to NaN
        if q5 == "null":
            recruit.iloc[row_index, 14] = "NaN"

# axis = 0 refers to apply acting on rows
recruit.apply(determine_Q5, axis=0) 
```

```python
# Q12 removed
# Q13 is a choose all that apply question so each answer results in its own column
recruit.columns.values[20] = "Q13_A"
recruit.columns.values[21] = "Q13_B"
recruit.rename(columns={'Women who identify as Black or African American':'Q13_C1',
    'Women who identify as Latina':'Q13_C2',
    'Women who identify as Asian':'Q13_C3',
    'Women who identify as Native American/Alaskan/Pacific Islander':'Q13_C4',
    'Women who identify as Mixed Race':'Q13_C5',
    'Women who identify as White/Caucasian':'Q13_C6',
    'Women who identify as Disabled':'Q13_G7',
    'Women who identify as LGBTQ':'Q13_C8',
    'Persons who identify as gender non-binary':'Q13_C9',
    'None':'Q13_J'
                  },inplace=True)
```

Rename the following questions: Q14 through Q21. Q16 skipped, since removed.

```python
recruit.columns.values[32] = "Q14"
recruit.columns.values[33] = "Q15"
recruit.columns.values[34] = "Q17"
recruit.columns.values[35] = "Q18"
recruit.columns.values[36] = "Q19"
recruit.columns.values[37] = "Q20_A"
recruit.columns.values[38] = "Q20_B"
recruit.columns.values[39] = "Q20_C"
recruit.columns.values[40] = "Q20_D"
recruit.columns.values[41] = "Q21"
recruit
```

```python
def determine_Q15(row):
    for row_index in range(recruit.shape[0]):
        q15 = recruit.iloc[row_index, 33]
        if q15 == "No":
            #Q17, 18, 19 all N/A: Q16 already removed
            recruit.iloc[row_index, 34] = "N/A"
            recruit.iloc[row_index, 35] = "N/A"
            recruit.iloc[row_index, 36] = "N/A" 
       
# axis = 0 refers to apply acting on rows
recruit.apply(determine_Q15, axis=0) 
```

```python
#Q19 is a choose all that apply question so each answer results in its own column
recruit.rename(columns={'At the secondary or tertiary education level':'Q19_A',
    'At the on-ramping/workforce re-entry level for mid-career women':'Q19_B',
    'No':'Q19_C',
    'Other':'Q19_D'},inplace=True)
```

```python
#### Adding the benchmark column

#Groups are: Financial Services, Technology, General, Non-profit
# A, B, F, J, Q, U, V financial
# C, G, M, P General
# D, H, N, O, R, S, T, X Technology
# E, I, K, L, W Non-Profit

financial_group = ["Company 1A","Company 1B","Company 1F","Company 1J","Company 1Q","Company 1U","Company 1V"]
technology_group = ["Company 1D","Company 1H","Company 1N","Company 1O","Company 1R","Company 1S","Company 1T","Company 1X"]
nonprofit_group = ["Company 1E","Company 1I","Company 1K","Company 1L","Company 1W"]

#Creating new column to contain benchmark group info
recruit["benchmark_group"] = np.nan

#Resetting benchmark group for finance companies 
finance = recruit["Organization"].isin(financial_group)
recruit.loc[finance,"benchmark_group"] = "Finance"

#Resetting benchmark group for technology companies 
technology = recruit["Organization"].isin(technology_group)
recruit.loc[technology,"benchmark_group"] = "Technology"

#Resetting benchmark group for technology companies 
nonprofit = recruit["Organization"].isin(nonprofit_group)
recruit.loc[nonprofit,"benchmark_group"] = "NonProfit"


#Printing out the dataframe Organization and benchmark columns to double check
#Note that any companies in general will remain as Nan as these companies visualizations
#will not be shown with a benchmark filter
recruit[["Organization","benchmark_group"]]
```
